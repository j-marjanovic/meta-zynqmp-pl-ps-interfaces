# AXI Proxy on repeat

In the previous notebook ([01-axi-proxy.ipynb](./01-axi-proxy.ipynb)) we have performed AXI transactions using AXI Proxy. We have used an external program, which has caused that between each invocation of the program, the state of the L1 and L2 caches has significantly changed.

If we imagine an application where the program runs continously, and we use one of AXI ports to exhange the data between the SW and the FPGA, a more realistic measurement would be to perform the measurements repeatedly.

In this notebook we use a Python implementation of the "driver" for the AXI Proxy (from the library `zynqmp_pl_ps`)

```python
import logging
import os
import random
import re
import subprocess
from dataclasses import dataclass
from typing import Dict

import matplotlib.pyplot as plt
import numpy as np
from fpga_mngr_interface.FpgaManagerInterface import FpgaManagerInterface
from zynqmp_pl_ps import AxiProxy, Udmabuf, UioDev
```

## Programming the FPGA

We check if the FPGA is programmed, and if that is not the case, we program it with the appropriate bitstream and device tree overlay.

```python
# logging.basicConfig(level=logging.DEBUG)

fmi = FpgaManagerInterface()
is_programmed = fmi.is_programmed()

try:
    UioDev.get_uio_dev_file("AxiProxy", "hp")
    ap_found = True
except RuntimeError:
    ap_found = False

print(f"is programmed: {is_programmed}, AXI Proxy found: {ap_found}")

if not is_programmed or not ap_found:
    print("Re-programming the FPGA")
    fmi.remove()
    fmi.program_bitstream(
        "/lib/firmware/xilinx/axi_proxy/axi_proxy.bit.bin",
        "/lib/firmware/xilinx/axi_proxy/axi_proxy.dtbo",
    )
```

## Test runner

```python
@dataclass(frozen=True)
class TestConfig:
    port_name: str
    udmabuf_flags: int
    axi_cache: int
    axi_prot: int
    axi_user: int
```

```python
#! /usr/bin/env python3

import random
import time


class AxiProxyTest:
    OFFS = 0x100

    def __init__(self, test_config: TestConfig):
        dev = UioDev.get_uio_dev_file("AxiProxy", search_note=test_config.port_name)
        self.ap = AxiProxy.AxiProxy(dev)

        self.udmabuf = Udmabuf.Udmabuf("axi:udmabuf@0x0", test_config.udmabuf_flags)
        assert self.udmabuf._get_value("dma_coherent") == 1

        self.ap.config_axi(
            cache=test_config.axi_cache,
            prot=test_config.axi_prot,
            user=test_config.axi_user,
        )

        self.rnd = random.Random(123)

    def hw_write(self):
        data = self.rnd.randint(0, 2 ** 32 - 1)
        duration = self.ap.write(self.udmabuf.phys_addr + self.OFFS, [data, 0, 0, 0])
        print(f"AxiProxy: write data = {data:08x}, duration = {duration}")

        return data, duration

    def hw_read(self):
        data_rbv, duration = self.ap.read(self.udmabuf.phys_addr + self.OFFS)
        print(f"AxiProxy: read data  = {data_rbv[0]:08x}, duration = {duration}")

        return data_rbv[0], duration
    
    def sw_write(self):
        data = self.rnd.randint(0, 2 ** 32 - 1)
        self.udmabuf.wr32(self.OFFS, data)
        print(f"Udmabuf : write data = {data:08x}")

        return data

    def sw_read(self):
        data_rd = self.udmabuf.rd32(self.OFFS)
        print(f"Udmabuf : read data  = {data_rd:08x}")

        return data_rd
```

## Measurement configuration

```python
udmabuf = Udmabuf.Udmabuf("axi:udmabuf@0x0")
BUFFER_SIZE = udmabuf.buf_size
print(f"Buffer size: {BUFFER_SIZE}")
del udmabuf
```

```python
NR_MEAS_PER_PORT = 100

CONFIGS = {
    "hp": TestConfig("hp", os.O_SYNC, axi_cache=0, axi_prot=0, axi_user=0),
    "hpc": TestConfig("hpc", 0, axi_cache=15, axi_prot=2, axi_user=1),
    "acp": TestConfig("acp", 0, axi_cache=15, axi_prot=2, axi_user=1),
}
```

## Measurement

```python
np.random.seed(1234)

meas = dict()

for interface, config in CONFIGS.items():

    test = AxiProxyTest(config)

    durs_rd = []
    durs_wr = []

    print("==================================")
    print(interface)

    for i in range(NR_MEAS_PER_PORT):
        data_hw_wr, duration = test.hw_write()
        durs_wr.append(duration)
        data_sw_rd = test.sw_read()
        assert data_hw_wr, data_sw_rd

        data_sw_wr = test.sw_write()
        data_hw_rd, duration = test.hw_read()
        durs_rd.append(duration)
        assert data_hw_wr, data_sw_rd

    meas[interface] = dict()
    meas[interface]["rd"] = durs_rd
    meas[interface]["wr"] = durs_wr
```

## Plotting

```python
@dataclass
class MeasData:
    hp: np.array
    hpc: np.array
    acp: np.array


def process_meas(data: dict, tx_type: str):
    hp = np.array(data["hp"][tx_type])
    hpc = np.array(data["hpc"][tx_type])
    acp = np.array(data["acp"][tx_type])

    return MeasData(hp=hp, hpc=hpc, acp=acp)


def plot_graph_hist(
    ax1, ax2,
    meas_hp: np.array,
    meas_hpc: np.array,
    meas_acp: np.array,
    title: str,
):
    ax1.plot(meas_hp, label="HP")
    ax1.plot(meas_hpc, label="HPC")
    ax1.plot(meas_acp, label="ACP")
    ax1.grid(True)
    ax1.legend(loc="upper right")
    ax1.set_ylabel("delay [clk cycles]")
    ax1.set_xlabel("iteration")

    delay_max = np.max([np.max(meas_hp), np.max(meas_hpc), np.max(meas_acp)])
    delay_min = np.min([np.min(meas_hp), np.min(meas_hpc), np.min(meas_acp)])

    HIST_CONF = {
        "orientation": "horizontal",
        "alpha": 0.7,
    }
    ax2.hist(meas_hp, label="HP", bins=30, range=(delay_min, delay_max), **HIST_CONF)
    ax2.hist(meas_hpc, label="HPC", bins=30, range=(delay_min, delay_max), **HIST_CONF)
    ax2.hist(meas_acp, label="ACP", bins=30, range=(delay_min, delay_max), **HIST_CONF)
    ax2.grid(True)
    ax2.legend(loc="upper right")
    ax2.set_xlabel("counts")

    ax1.set_title(title)
    ax2.set_title(title + " (hist)")
```

```python
data_read = process_meas(meas, "rd")
data_write = process_meas(meas, "wr")
```

```python
fig = plt.figure(figsize=(12, 10))
[ax11, ax12], [ax21, ax22] = fig.subplots(
    2, 2, gridspec_kw={"width_ratios": [3, 1]}, sharey="row"
)


plot_graph_hist(
    ax11,
    ax12,
    data_write.hp,
    data_write.hpc,
    data_write.acp,
    "Write latency",
)
plot_graph_hist(
    ax21,
    ax22,
    data_read.hp,
    data_read.hpc,
    data_read.acp,
    "Read latency",
)

fig.suptitle("AXI Proxy on repeat, clk freq = 250 MHz")
```

## Conclusion

* in this more realistic scenario we obtain better results
* the latency of the ACP port is extremelly low and very constant
* similarly as in the measurement in the previous notebook, the writes are faster on HP port and reads are faster on the HPC block
* we see that the first write to ACP takes more time; this likely because the L1 cache needs to check if the value is in L2 cache
* on the contrary, the first write on the HPC port is faster than all subsequent writtes; this is likely because the cache line is invalid, and the CCI does not need to communicate with the APU
